```python
import matplotlib.pyplot as plt
# 막대그래프 그리기
plt.bar([1,2,3,4], [1,4,9,16])
plt.xlabel("score")
plt.ylabel("grade")
plt.show()
```

```python
# 곡선 그래프 그리기
plt.plot([1,2,3,4], [2.0, 3.0, 5.0, 10.0], 'r')
plt.plot([1, 2, 3, 4], [2.0, 2.8, 4.3, 6.5], 'g')
plt.plot([1, 2, 3, 4], [2.0, 2.5, 3.3, 4.5], 'b')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.show()
```
